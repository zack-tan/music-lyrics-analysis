## Defining functions to acces lyric pools



```python
#getting artist lyric pool by ID across ALL data

def artistLyricPool(artist_id, df):
    #getting artist song data
    lyrics_artist = df.loc[df['artist_id'] == artist_id]['lyrics_clean_stemmed']
    #aggregating lyrics
    i=0
    for song in lyrics_artist:

        if i == 0:
            lyrics = song
        else :
            lyrics += song
        i+=1
    return lyrics
```

```python
#getting artist lyric pool by ID for a specific year
def artistLyricPoolByYear(artist_id, df, year):
    #getting artist song data
    lyrics_artist = df.loc[df['artist_id'] == artist_id][['lyrics_clean_stemmed', 'track_release_year']]
    lyrics_artist_year = lyrics_artist.loc[lyrics_artist['track_release_year'] == year]['lyrics_clean_stemmed']
    #aggregating lyrics
    i=0
    for song in lyrics_artist_year:

        if i == 0:
            lyrics = song
        else :
            lyrics += song
        i+=1
    return lyrics
```

## Training the LDA

Credit where credit is due !
https://www.kaggle.com/ktattan/lda-and-document-similarity proved to be really helpful, the methods below are taken off @ktattan

```python
import gensim
import time
from gensim.models import LdaModel
from gensim import models, corpora, similarities

def train_lda(data):
    """
    This function trains the lda model
    We setup parameters like number of topics, the chunksize to use in Hoffman method
    We also do 2 passes of the data since this is a small dataset, so we want the distributions to stabilize
    """
    #this is where you specify number of topics
    num_topics = 100
    chunksize = 300
    dictionary = corpora.Dictionary(data['Lyrix_Pool'])
    corpus = [dictionary.doc2bow(doc) for doc in data['Lyrix_Pool']]
    t1 = time.time()
    # low alpha means each document is only represented by a small number of topics, and vice versa
    # low eta means each topic is only represented by a small number of words, and vice versa
    lda = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary,
                   alpha=1e-2, eta=0.5e-2, chunksize=chunksize, minimum_probability=0.0, passes=2)
    t2 = time.time()
    print("Time to train LDA model on ", len(data), "articles: ", (t2-t1)/60, "min")
    return dictionary,corpus,lda
```

```python
#train lda
#https://www.kaggle.com/ktattan/lda-and-document-similarity

#Initial train - total corpus
dictionary,corpus, lda =train_lda(rapperpools)
```

## Defining LDA query result calculation

```python
from scipy.stats import entropy
def jensen_shannon(query, matrix):
    """
    This function implements a Jensen-Shannon similarity
    between the input query (an LDA topic distribution for a document)
    and the entire corpus of topic distributions.
    It returns an array of length M where M is the number of documents in the corpus
    """
    # lets keep with the p,q notation above
    p = query[None,:].T + np.zeros([100, 876])# take transpose
    q = matrix.T # transpose matrix
    m = 0.5*(p + q)
    return np.sqrt(0.5*(entropy(p,m) + entropy(q,m)))
```

```python
def get_most_similar_documents(query,matrix,k=10):
    """
    This function implements the Jensen-Shannon distance above
    and retruns the top k indices of the smallest jensen shannon distances
    """
    sims = jensen_shannon(query,matrix) # list of jensen shannon distances
    
    ind = np.argpartition(sims, -k)[-k:]
    
    return sims.argsort()[:k], sims[ind] # the top k positional index of the smallest Jensen Shannon distances
```

## Time-specific query

```python

def MatchMyArtist(rapper_name, compare_to_years, compare_against_years):
    #create my rapper lyric pool for my desired year range 
    myrapper_pool = []
    for year in compare_to_years:
        if artistLyricPoolByYearByName(rapper_name, hiphop, year):
            myrapper_pool += artistLyricPoolByYearByName(rapper_name, hiphop, year)
        
    #setting up my pool to be used in my upcoming LDA model    
    my_rapper_bow = dictionary.doc2bow(myrapper_pool)
    
    #create the lyric pool of all rappers for desired year range to compare my rapper pool to !
    compare_against_df = hiphop[hiphop.track_release_year.isin(compare_against_years)] 
    print(len(compare_against_df))
    
    
    compare_against_rapper_names=compare_against_df['artist_name'].unique()
    compare_against_pools = []
    
    for name in compare_against_rapper_names:
        lyrix_pool = artistLyricPoolByName(name,compare_against_df)
        compare_against_pools.append(lyrix_pool)
    
    
    #reacreating pools dataset
    againstpools = pd.DataFrame({'artist_name':compare_against_rapper_names, 'Lyrix_Pool':compare_against_pools})
    #adding bow data
    
    df2 = {'artist_name': 'Drake', 'Lyrix_Pool': myrapper_pool}

    againstpools = againstpools.append(df2, ignore_index = True)

    

    
    #training the lda
    against_dictionary,against_corpus, against_lda = train_lda(againstpools)
    against_doc_topic_dist = np.array([[tup[1] for tup in lst] for lst in against_lda[against_corpus]])
    #obtaining the prob distribution in the new document
    to_doc_distribution = np.array([tup[1] for tup in against_lda.get_document_topics(bow = my_rapper_bow)])
    
    #defining js distance computation
    def jensen_shannon_1(query, matrix):
       
        # lets keep with the p,q notation above
        p = query[None,:].T + np.zeros([100, len(againstpools)])# take transpose
        q = matrix.T # transpose matrix
        m = 0.5*(p + q)
        return np.sqrt(0.5*(entropy(p,m) + entropy(q,m)))
    
    def get_most_similar_documents_1(query,matrix,k=10):
   
        sims = jensen_shannon_1(query,matrix) # list of jensen shannon distances
    
        ind = np.argpartition(sims, -k)[-k:]
    
        return sims.argsort()[:k], sims[ind] # the top k positional index of the smallest Jensen Shannon distances
    
    #fetching best matches
    most_sim_ids, most_sim_id_values = get_most_similar_documents_1(to_doc_distribution,against_doc_topic_dist)
    most_similar_df = againstpools[againstpools.index.isin(most_sim_ids)]
    most_similar_df["similarity_score_js"] = most_sim_id_values
    most_similar_df = most_similar_df[most_similar_df.artist_name != rapper_name]
    most_similar_df = most_similar_df.sort_values(by='similarity_score_js', ascending=True)
    
    
    return most_similar_df[['artist_name', 'similarity_score_js']]
   
```
